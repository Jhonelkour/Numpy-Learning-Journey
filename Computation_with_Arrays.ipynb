#### Other Operations On Arrays 
##### Other Methods and function 

```python
# Mathematical functions :
a = np.array([[1,2,3,4],
              [5,6,7,8]])
```

#### np.sum()

```python
# Computes the sum of all elements in the array.
np.sum(a)  # 1+2+3+4+5+6+7+8 = 36 

# np.sum(a, axis=0)  # sum down each column → [6, 8, 10, 12]
# np.sum(a, axis=1)  # sum across each row → [10, 26]
```

#### np.prod()

```python
# Computes the product of all elements in the array.
np.prod(a) # # 1×2×3×4×5×6×7×8 = 40320 

# np.prod(a, axis=0)  # [5, 12, 21, 32]
# np.prod(a, axis=1)  # [24, 1680]
```

#### np.min() and np.max()

```python
# Returns the minimum or maximum value.
np.min(a)  # 1
np.max(a)  # 8

# np.min(a, axis=0)  # [1, 2, 3, 4]
# np.max(a, axis=1)  # [4, 8]

```

#### np.argmin() and np.argmax()

```python
# Returns the index (in the flattened array) of the min/max value.

np.argmin(a)  # 0 (value = 1)
np.argmax(a)  # 7 (value = 8)

# np.argmin(a, axis=1)  # [0, 0] (min value index in each row)
# np.argmax(a, axis=0)  # [1, 1, 1, 1] (max index in each column) 

```

#### np.ptp()

```python
# Returns the range of values: max - min (called "peak to peak").

np.ptp(a)  # 8 - 1 = 7

# np.ptp(a, axis=1)  # [3, 3] (row-wise)
# np.ptp(a, axis=0)  # [4, 4, 4, 4] (column-wise)
```

#### Statistics 

##### np.mean()

```python
# Calculates the average of all elements.

np.mean(a)  # (1+2+3+4+5+6+7+8)/8 = 4.5

# np.mean(a, axis=0)  # [3., 4., 5., 6.]   column-wise means
# np.mean(a, axis=1)  # [2.5, 6.5]         row-wise means
```

##### np.std() — Standard Deviation

```python
# Measures the spread of the data from the mean.

np.std(a)  # ≈ 2.291

# np.std(a, axis=0)  # [2., 2., 2., 2.]  → std per column
# np.std(a, axis=1)  # [1.118, 1.118]    → std per row
```

##### np.var() — Variance

```python
# Square of the standard deviation (measures spread).

np.var(a) # ≈ 5.25 

# np.var(a, axis=0)  # [4., 4., 4., 4.]
# np.var(a, axis=1)  # [1.25, 1.25]
```

#### Truth value testing 

##### np.any()

```python
# Returns True if at least one element is truthy (non-zero).
a = np.array([0, 0, 3])
if np.any(a):
    print('At least one no-zeros')
```

##### np.all()

```python
b = np.array([1,2,3,4])
if np.all(b):
    print('All are non-zeros')
```

```python
# a > 2  # Returns an array: [False, False, True]
(a > 2).any()  # True if any element > 2
```

#### a.any() and a.all() as methods

```python
a.any()

# Checks if at least one element is non-zero (truthy)
# In this case: 3 is non-zero 
# Result True
```

```python
a.all()

# Checks if all elements are non-zero
# Here: 0, 0, 3  Two zeros 
# Result False
```

```python
b.any()

# Checks if all elements are non-zero (truthy)
# Since all elements are ≠ 0:
# Result True
```

```python
b.all()

# Checks if at least one element is non-zero
# Since all are non-zero, this is definitely true:
# Result → True
```

#### Unraveling 

```python
# Create a 2D array with shape (2, 3)
a = np.array([[10,20,30],
              [40,50,60]])
np.unravel_index(a.argmax(), a.shape)
```

#### Explication 

```python
# a.argmax() returns the index of the maximum value in the flattened array
# Flattened: [10, 20, 30, 40, 50, 60]
# Max value is 60, which is at index 5 (0-based)
max_index_flat = a.argmax()  #  5

# np.unravel_index converts a flat index (from a 1D view) back to a 2D index based on the shape of `a`
# a.shape = (2, 3), so unravel_index(5, (2, 3)) gives the position (1, 2)
max_position = np.unravel_index(max_index_flat, a.shape)  # ➞ (1, 2)

# Print results
print("Max value:", a.max())                     # 60
print("Flat index of max:", max_index_flat)      # 5
print("2D index of max:", max_position)          # (1, 2)
```
